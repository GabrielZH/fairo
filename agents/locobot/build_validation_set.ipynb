```python
# build validation set
import random
import os
import matplotlib.pyplot as plt
import cv2
import shutil

reex_dir = '/home/locobotm/0512_data/reexplore_final/'

dst = os.path.join(reex_dir, 'val_set_sample')
if os.path.isdir(dst):
    print(f'rmtree {dst}')
    shutil.rmtree(dst)

os.makedirs(dst, exist_ok=True)
os.makedirs(os.path.join(dst, 'rgb'), exist_ok=True)

valid_objs = [0, 2, 3, 4]
heu = ['c1pp', 'r1', 's1pp']

val_set_samples = []

# random.seed(7)

for x in valid_objs:
    for h in heu:
        hdir = os.path.join(reex_dir, str(x), h)
        
        # get a random sample and display it
        c = random.choice(os.listdir(os.path.join(hdir, 'rgb')))
        src_img = cv2.imread(os.path.join(hdir, 'rgb', c))
        print(f'obj {x}, heu {h}, total = {len(val_set_samples)+1}')
        plt.imshow(src_img[...,::-1])
        plt.show()
        
        shutil.copyfile(os.path.join(hdir, 'rgb', c), os.path.join(dst, 'rgb', f'{len(val_set_samples):05d}.jpg'))
        
        val_set_samples.append(c)
```
