```python
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
robot_frame_filenames = set([fname.split("_")[0] 
                             for fname in glob.glob("robotframes/*")])
habitat_frame_filenames = set([fname.split("_")[0] 
                               for fname in glob.glob("habitatframes/*")])
```

```python
def add_noise_to_depth(depth, gamma_shape=1000, gamma_scale=0.001):
    """
    Arguments:
        depth: depth image in [min_depth, max_depth]
    """
    depth = depth.copy()
    
    # TODO How to handle values at range boundaries (min_depth, max_depth)?
    
    # Multiplicative noise: Gamma random variable
    multiplicative_noise = np.random.gamma(gamma_shape, gamma_scale)
    depth = multiplicative_noise * depth
    
    # TODO How to add additive noise on depth image instead of point cloud?

    return depth

def visualize(fname, max_depth=4.0):
    rgb = np.load(f"{fname}_rgb.npy")
    depth = np.load(f"{fname}_depth.npy")
    noisy_depth = add_noise_to_depth(depth)
    
    plt.figure()
    _, subplots = plt.subplots(1, 3)
    subplots[0].imshow(rgb)
    subplots[0].set_axis_off()
    subplots[1].imshow(depth)
    subplots[1].set_axis_off()
    subplots[2].imshow(noisy_depth)
    subplots[2].set_axis_off()
    plt.show()
    
    print(f"Depth range: ({depth.min()}, {depth.max()})")
    print(f"Noisy depth range: ({noisy_depth.min()}, {noisy_depth.max()})")
    print("_" * 40)
```

```python
for fname in robot_frame_filenames:
    visualize(fname)
```

```python
for fname in habitat_frame_filenames:
    visualize(fname)
```
